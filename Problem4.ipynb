```python
import pandas as pd
import numpy as np
import math
import statistics
```

```python
marks = [99,45,60,80,95,100,95,91,85,87,77,75,61,71,85,88,83,85,79,81,82,55,63,75,82,88
,77,78,41,70]
```

```python
mean_marks = statistics.mean(marks)
```

```python
#the mean of the given dataset of marks
mean_marks
```

```python
stdev_marks = statistics.stdev(marks)
#the standard deviation of marks
stdev_marks
```

```python
sample_variance = stdev_marks**2
```

```python
#the sample variance of the given dataset of marks
sample_variance
```

```python
#maximum likelihood method
log_marks = []
for i in range(len(marks)):
  log_marks.append(math.log(marks[i]))
log_marks
```

```python
#the mean of all the logvalues of the marks dataset
mean_logvals = statistics.mean(log_marks)
mean_logvals
```

```python
# finding the lambda parameter using maximum likelihood method
lambda_marks = mean_logvals # we found lambda parameter
```

```python
#process for finding the zeta parameter using maximum likelihood method
zeta = 0 
for i in range(len(log_marks)):
  zeta += pow((log_marks[i] - mean_logvals),2)
zeta = zeta/30
zeta = pow(zeta, 0.5)
zeta # we found another parameter zeta 
```

```python
print(f"Parameters of lognormal distribution using maximum likelihood method are lambda : {lambda_marks}  and zeta : {zeta}")
```

```python
#Method of moments
zeta_mom = math.log((sample_variance/pow(mean_marks,2))+1)
zeta_mom = pow(zeta_mom,0.5)
```

```python
#this is the zeta value found by using method of moments
zeta_mom
```

```python
#this is the lambda value found by using method of moments
lambda_mom = math.log(mean_marks) - (0.5*pow(zeta_mom,2))
lambda_mom
```

```python
print(f"The parameters of lognormal distribution using method of moments are lambda : {lambda_mom} and zeta : {zeta_mom}")
```
