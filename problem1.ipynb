```python
import pandas as pd
import numpy as np
import math

from scipy.stats import chi2_contingency
from scipy.stats import chi2
```

```python
rainfall_freq = [57,30,9,3,1,0,0,0]
rainfall_data = pd.DataFrame(rainfall_freq, columns = ["Frequency"])
rainfall_data
```

```python
def nCr(n,r):
  f = math.factorial
  return f(n)//(f(r)*f(n-r))
  
```

```python
#alpha and probability values
alpha  = 0.05
prob = 1 - alpha
```

```python
#f is the dof 
f = 5
# here we are finding out the value of c with the given f and probability values 
c = chi2.ppf(prob, f)
c
```

```python
#these are all the expected values of rainfall frequency using binomial distribution
expected  = []
for i in range(0,8):
  expected.append(nCr(7,i)*pow(0.1,i)*pow(0.9,7-i)*100)
```

```python
expected
```

```python
#we are finding the error between expected and observed values
error = 0
for i in range(0,8):
  error += (pow((expected[i] - rainfall_freq[i]),2)/expected[i])
error
```

```python
if error<c:
  print("Our null hypothesis is correct. That is we can distribute this model as a binomial distribution B(M = m|(7,0.1))")
else:
  print("our null hypothesis is wrong")
```
