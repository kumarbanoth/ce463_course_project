```python
import pandas as pd
import numpy as np
import math

from scipy.stats import chi2_contingency
from scipy.stats import chi2
```

```python
rainfall_freq = [57,30,9,3,1,0,0,0]
rainfall_data = pd.DataFrame(rainfall_freq, columns = ["Frequency"])
rainfall_data
```

```python
# we are defining a dn-alpha value table
from scipy.stats import ksone

def ks_critical_value(n_trials, alpha):
    return ksone.ppf(1-alpha/2, n_trials)
```

```python
dn_ks = ks_critical_value(8,0.05)
dn_ks
```

```python
def nCr(n,r):
  f = math.factorial
  return f(n)//(f(r)*f(n-r))
  
```

```python
#the alpha and probability values
alpha  = 0.05
prob = 1 - alpha
```

```python
#these are all the expected values using binomial distribution
expected  = []
for i in range(0,8):
  expected.append(nCr(7,i)*pow(0.1,i)*pow(0.9,7-i))
expected
```

```python
#in this code we are finding the CDF of observed and expected values 
cum_sn = rainfall_freq.copy()
cum_sn
for i in range(len(rainfall_freq)):
  cum_sn[i] = cum_sn[i]/100
  if i>0:
    cum_sn[i] = cum_sn[i-1] + cum_sn[i]


for i in range(len(expected)):
  if i>0:
    expected[i] = expected[i-1] + expected[i]

expected
```

```python
#these are the expected and observed CDF values
cum_sn, expected
```

```python
#here we ar finding the dn value which is the maximum of the error between observed and expected values
error = []
for i in range(len(cum_sn)):
  if cum_sn[i]>=expected[i]:
    error.append(cum_sn[i] - expected[i])
  else:
    error.append(expected[i] - cum_sn[i])

dn = max(error)
```

```python
dn, dn_ks
```

```python
if dn<dn_ks:
  print("Our null hypothesis is correct")
else:
  print("Our null hypothesis is wrong")
```
