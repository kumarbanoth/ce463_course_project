```python
import pandas as pd
import numpy as np
import math
import statistics
```

```python
n = 9
pile_capacities = [82,75,95,90,88,92,78,85,80]
piles = pd.DataFrame(pile_capacities, columns = ["Pile capacity"], index = [1,2,3,4,5,6,7,8,9])
piles
```

```python
#calculating the mean and standard deviation of the given sample
mean_sample = statistics.mean(pile_capacities)
stdev_sample = statistics.stdev(pile_capacities)
```

```python
print(f"The mean of our test pile capacities is {mean_sample}")
print(f"The standard deviation of our test pile capacities is {round(stdev_sample,4)}")
```

```python
#the alpha and probability values
alpha = 0.05
prob = 1 - alpha
```

```python
alpha, prob
```

```python
#We have to use t-test because we don't know the variance
t= (mean_sample - 80)*pow(9,0.5)/stdev_sample
t
```

```python
#this is the t-alpha value for 5% significance level
t_alpha = 2.306 
```

```python
if t<t_alpha:
  print("Null hypothesis is correct")
else:
  print("Null hypothesis is wrong")
```

```python
#99% confidence interval 
lower_limit = (mean_sample) - (2.58*stdev_sample)/pow(9,0.5)
upper_limit = (mean_sample) + (2.58*stdev_sample)/pow(9,0.5)
lower_limit, upper_limit
```

```python
confidence_interval = [lower_limit, upper_limit]
```
